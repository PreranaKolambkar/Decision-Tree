# Treating outliers

```python
import scipy.stats as stats
```

```python
def find_outliers_with_extremes(df,column):
    Q1 = df[column].quantile(0.25)   # find Q1
    Q3 = df[column].quantile(0.75)   #Find Q2
    IQR = Q3 - Q1                    # Find IQR
    lower_bound = Q1 - 1.5 * IQR     # calcualte the lower bound
    upper_bound = Q3 + 1.5 * IQR     # calcualte the upper bound
    print(f"Lower_bond: {lower_bound}")
    print(f"Upper_bond: {upper_bound}") 
    
    outlier_rows = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    # find the rows which has outlierd
    return outlier_rows,lower_bound,upper_bound  # return the value
```

# Decision Tree Classification:

```python
target.shape
```

```python
df5=pd.merge(df_new1,target,left_index=True,right_index=True)
# to consider the target value for the features present
```

```python
df5.shape
```

```python
target_new=df5[['num']]
target_new.shape
```

```python
features_new=df5.drop(columns='num')
features_new.shape
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(features_new,target_new,train_size=0.8,random_state=150)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dec_tree=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=3)
```

```python
dec_tree.fit(x_train,y_train)
```

```python
ypred=dec_tree.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
#i) max_depth: it is maximum depth of tree

#A smaller value helps prevent overfitting by restricting the tree’s complexity.

#A larger value allows the tree to capture more complex patterns but increases the risk of overfitting.
```

```python
#ii)criterion:he function to measure the quality of a split

#gini or entropy

#gini is faster to compute and is the default in scikit-learn.
#entropy can be more sensitive to informative splits but is computationally heavier.

```

```python
#splitter: The strategy used to choose the split at each node.
# best (chooses the best split) or random (chooses a random split).
#best ensures the highest information gain or Gini reduction.
#random can lead to faster training but might result in less optimal trees.
```
